```python
def log_transform(df):
    features_logged = np.log1p(df)
    return features_logged

def normalize_data(df):
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(df)
    return pd.DataFrame(features_scaled, columns=df.columns, index=df.index)

def min_max_scaler(df):
    scaler = MinMaxScaler(feature_range=(0,1))
    features_scaled = scaler.fit_transform(df)
    return pd.DataFrame(features_scaled, columns=df.columns, index = df.index)

def one_scaler(df):
    df_scaled = df.div(df.sum(axis=1), axis=0)
    return df_scaled
```

```python
sample_ids = balanced_df.iloc[:, 0]        # ID
labels = balanced_df.iloc[:, -1]           # label
features = balanced_df.iloc[:, 1:-1]       # features
```

```python
#df_logged = log_transform(features) #dont log transform
df_norm = one_scaler(features)
df_norm = normalize_data(df_norm) #z normalization


#df_norm = min_max_scaler(features)
#rescale
```
